```python
import weaviate
import openai
```

```python
openai.api_key = ""
embeddings = openai.Embedding.create(input=["danilo rojas"],model="text-embedding-3-small")
```

```python
client = weaviate.Client(
    url="http://localhost:8080",
    additional_headers = {
        "X-OpenAI-Api-Key": ""  # Replace with your inference API key
    }
)
```

```python
embeddings["data"][0]["embedding"]
```

```python
response = (client.query.get("DocketDocuments",["name","storageKey","originalPath","size","iD","content"])
.with_near_text({
    "concepts": ["SHEETCHOLINE"]
}).with_limit(5).with_additional(["distance"]).do())
```

```python
response = (client.query.get("DocketDocuments",["name","storageKey","originalPath","size","iD","content"])
.with_hybrid(query="SHEETCHOLINE").with_limit(5).with_additional(["distance"]).do())
```

```python
response = (client.query.get("DocketDocuments",["name","storageKey","originalPath","size","iD","content"])
.with_near_vector({
    "vector": embeddings["data"][0]["embedding"],
}).with_limit(5).with_additional(["distance"]).do())
```

```python
response
```

```python
for doc in response["data"]["Get"]["DocketDocuments"]:
    distance = doc["_additional"]["distance"] if doc["_additional"] else None
    name = doc["name"]
    content = doc["content"]
    print(f"Distance: {distance}, Name: {name}, Content: {content}")
```
